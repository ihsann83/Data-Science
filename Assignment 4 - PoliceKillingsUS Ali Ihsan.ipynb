## 2: Plot 12 most dangerous cities

```python
df = df.dropna(how="any")
most_danger=df["city"].value_counts()[:12]
most_danger


```

```python
df.head()
```

```python
# with seaborn barplot
```

```python
fig, ax = plt.subplots(figsize=(8,5))



sns.barplot(data=df, y=df["city"].value_counts()[:12].values, x=df["city"].value_counts()[:12].index, ax=ax);
ax.set_xticklabels(df["city"].value_counts()[:12].index, rotation=45)
ax.set_title("Most Dangerous Cities", color="blue", fontsize=15 );
```

```python
# with seaborn countplot
```

```python
plt.figure(figsize=(8,5))

sns.countplot(x='city', data=df, order=most_danger.index);
plt.title("Most Dangerous Cities", color="blue", fontsize=15 );
plt.xticks(rotation=45);
```

## 3: Plot 20 most dangerous states

```python
df["state"].value_counts()[:20]
```

```python
# with seaborn barplot
```

```python
fig, ax= plt.subplots(figsize=(7,5))

ax=sns.barplot(data=df, x=df["state"].value_counts()[:20].index, y=df["state"].value_counts()[:20].values, ax=ax)
ax.set_xticklabels(df["state"].value_counts()[:20].index, rotation=45)
ax.set_title("Most Dangerous States", color="blue", fontsize=15 );
plt.tight_layout()
```

```python
# with seaborn countplot
```

```python
plt.figure(figsize=(8,5))

sns.countplot(x='state', data=df, order=df["state"].value_counts()[:20].index);
plt.title("Most Dangerous States", color="blue", fontsize=15 );
plt.xticks(rotation=45);
```
